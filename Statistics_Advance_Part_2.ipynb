# **Statistics Advance Part 2 Theory Questions**

# 1. What is hypothesis testing in statistics?

Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It involves formulating two competing hypotheses and using sample data to determine which is more likely.

# 2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

* **Null Hypothesis (H₀):** Assumes no effect or no difference; it's the default or status quo.
* **Alternative Hypothesis (H₁ or Ha):** Represents the claim you want to test; indicates a significant effect or difference.

For example:

* H₀: μ = 50
* H₁: μ ≠ 50 (two-tailed) or μ > 50 / μ < 50 (one-tailed)

# 3. What is the significance level in hypothesis testing, and why is it important?

The **significance level (α)** is the probability of rejecting the null hypothesis when it is actually true (Type I error). Common values are 0.05, 0.01, and 0.10. It sets the threshold for deciding whether the observed result is statistically significant.

# 4. What does a P-value represent in hypothesis testing?

The **P-value** is the probability of obtaining results as extreme as the observed results, assuming the null hypothesis is true. A small P-value indicates strong evidence against the null hypothesis.

# 5. How do you interpret the P-value in hypothesis testing?

* If **P ≤ α**: Reject the null hypothesis (significant result)
* If **P > α**: Fail to reject the null hypothesis (not significant)

Example: P = 0.03, α = 0.05 → Reject H₀

# 6. What are Type 1 and Type 2 errors in hypothesis testing?

* **Type I Error (False Positive):** Rejecting H₀ when it is actually true (controlled by α).
* **Type II Error (False Negative):** Failing to reject H₀ when it is actually false (denoted by β).

# 7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

* **One-tailed test:** Tests for deviation in one direction (e.g., H₁: μ > μ₀).
* **Two-tailed test:** Tests for deviation in both directions (e.g., H₁: μ ≠ μ₀).

# 8. What is the Z-test, and when is it used in hypothesis testing?

The **Z-test** is used when:

* The population standard deviation is known.
* The sample size is large (n > 30).
  It's used to compare means, proportions, or to test a single mean against a population mean.

# 9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

**Z-score formula:**

$$
Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}
$$

Where:

* $\bar{x}$ = sample mean
* $\mu$ = population mean
* $\sigma$ = population standard deviation
* $n$ = sample size

It tells how many standard deviations a sample mean is from the population mean.

# 10. What is the T-distribution, and when should it be used instead of the normal distribution?

The **T-distribution** is used when:

* The population standard deviation is unknown.
* The sample size is small (n < 30).

It has heavier tails than the normal distribution, accounting for more variability in small samples.

---

# 11. What is the difference between a Z-test and a T-test?

The **main difference** between a Z-test and a T-test lies in the knowledge of population standard deviation and the sample size:

* Use a **Z-test** when:

  * The **population standard deviation** is known.
  * The **sample size is large** (typically n > 30).
  * The data is approximately normally distributed.

* Use a **T-test** when:

  * The **population standard deviation is unknown**.
  * The **sample size is small** (typically n < 30).
  * The data comes from a population that is approximately normal.

In essence, the T-test is a more flexible version of the Z-test used when there is more uncertainty (due to smaller sample size or unknown variability).

# 12. What is the T-test, and how is it used in hypothesis testing?

A **T-test** is used to compare means when population standard deviation is unknown. Types include:

* One-sample t-test
* Two-sample (independent) t-test
* Paired sample t-test

# 13. What is the relationship between Z-test and T-test in hypothesis testing?

Both tests compare sample means to population means but differ in assumptions:

* T-test is a generalization of Z-test used for small samples or unknown variance.
* As sample size increases, T-distribution approximates the normal (Z) distribution.

# 14. What is a confidence interval, and how is it used to interpret statistical results?

A **confidence interval (CI)** provides a range of plausible values for a population parameter. A 95% CI means that if the experiment is repeated many times, 95% of the CIs would contain the true parameter.

# 15. What is the margin of error, and how does it affect the confidence interval?

The **margin of error (MOE)** is the amount added/subtracted from the point estimate to create the CI:

$$
\text{CI} = \text{estimate} \pm \text{margin of error}
$$

Larger MOE = wider interval = less precision.

# 16. How is Bayes' Theorem used in statistics, and what is its significance?

**Bayes' Theorem** updates the probability of a hypothesis given new evidence:

$$
P(H|E) = \frac{P(E|H) \cdot P(H)}{P(E)}
$$

Used in fields like medical testing, spam filtering, and machine learning.

# 17. What is the Chi-square distribution, and when is it used?

The **Chi-square distribution** is used in:

* Testing independence between categorical variables.
* Testing goodness of fit of a model.
  It is skewed right and defined only for non-negative values.

# 18. What is the Chi-square goodness of fit test, and how is it applied?

This test checks if observed categorical data matches expected distribution:

$$
\chi^2 = \sum \frac{(O - E)^2}{E}
$$

Where $O$ = observed, $E$ = expected. If the value is too large, reject H₀.

# 19. What is the F-distribution, and when is it used in hypothesis testing?

The **F-distribution** is used to compare two variances and is the basis of ANOVA. It's right-skewed and depends on two degrees of freedom (numerator and denominator).

# 20. What is an ANOVA test, and what are its assumptions?

**ANOVA (Analysis of Variance)** tests if there are significant differences among group means.
Assumptions:

* Independence of observations
* Normality of residuals
* Homogeneity of variances (equal variances)

### 21. **What are the different types of ANOVA tests?**

* **One-way ANOVA:** One independent variable
* **Two-way ANOVA:** Two independent variables
* **Repeated Measures ANOVA:** Same subjects tested multiple times.

# 22. What is the F-test, and how does it relate to hypothesis testing?

An **F-test** compares the variances of two populations or models. It is used in:

* ANOVA (to test mean differences via variance)
* Regression analysis (to test overall model fit)

# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [51, 49, 52, 48, 50, 47, 53, 49, 50, 52]
population_mean = 50
population_std = 2
alpha = 0.05

# Calculate sample statistics
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Calculate Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate the p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Display results
print(f"Sample Mean      : {sample_mean:.2f}")
print(f"Z-Score          : {z_score:.3f}")
print(f"P-Value          : {p_value:.4f}")

# Interpretation
if p_value < alpha:
    print("Result: Reject the null hypothesis (significant difference).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference).")
```

# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# --- Step 1: Input Data ---
population_mean = 100          # H₀: μ = 100
population_std = 15            # Known population std dev
alpha = 0.05                   # Significance level (two-tailed)
sample = [105, 98, 110, 102, 95, 99, 108, 100, 97, 104]
sample_size = len(sample)

# --- Step 2: Z-test Calculation ---
sample_mean = np.mean(sample)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed

# --- Step 3: Print Results ---
print(f"Sample Mean       : {sample_mean:.2f}")
print(f"Z-Score           : {z_score:.3f}")
print(f"P-Value           : {p_value:.4f}")
print(f"Significance Level: {alpha}")

if p_value < alpha:
    print("Conclusion: Reject the null hypothesis (significant difference).")
else:
    print("Conclusion: Fail to reject the null hypothesis (no significant difference).")

# --- Step 4: Visualization ---
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

# Critical Z values
z_critical = norm.ppf(1 - alpha / 2)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Standard Normal Distribution', color='black')

# Shade rejection regions
plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.5, label='Rejection Region (left)')
plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.5, label='Rejection Region (right)')

# Plot the observed z-score
plt.axvline(z_score, color='blue', linestyle='--', linewidth=2, label=f'Observed Z = {z_score:.2f}')
plt.axvline(-z_critical, color='green', linestyle='--', label=f'Critical Z = ±{z_critical:.2f}')
plt.axvline(z_critical, color='green', linestyle='--')

plt.title("Two-Tailed Z-Test Visualization")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import matplotlib.pyplot as plt
import numpy as np
def visualize_errors(mu0, mu1, sigma, n, alpha):

  std_error = sigma / np.sqrt(n)

  # Calculate critical value(s) for a two-tailed test
  # For a two-tailed test, alpha is split into alpha/2 in each tail
  z_critical_lower = norm.ppf(alpha / 2, loc=mu0, scale=std_error)
  z_critical_upper = norm.ppf(1 - alpha / 2, loc=mu0, scale=std_error)

  # Calculate Beta (Type II error probability)
  # Beta is the probability of failing to reject H0 when H1 is true.
  # This is the area under the H1 distribution between the critical values.
  beta = norm.cdf(z_critical_upper, loc=mu1, scale=std_error) - norm.cdf(z_critical_lower, loc=mu1, scale=std_error)

  # Visualize the distributions and error regions
  x = np.linspace(min(mu0, mu1) - 4 * std_error, max(mu0, mu1) + 4 * std_error, 1000)
  pdf_h0 = norm.pdf(x, loc=mu0, scale=std_error)
  pdf_h1 = norm.pdf(x, loc=mu1, scale=std_error)

  plt.figure(figsize=(12, 6))
  plt.plot(x, pdf_h0, label=f'H0: μ = {mu0}', color='blue')
  plt.plot(x, pdf_h1, label=f'H1: μ = {mu1}', color='green')

  # Shade Type I error (Alpha) - area under H0 in rejection regions
  plt.fill_between(x, pdf_h0, where=(x <= z_critical_lower), color='red', alpha=0.5, label=f'Type I Error (α/2 = {alpha/2:.3f})')
  plt.fill_between(x, pdf_h0, where=(x >= z_critical_upper), color='red', alpha=0.5)

  # Shade Type II error (Beta) - area under H1 in acceptance region
  plt.fill_between(x, pdf_h1, where=((x >= z_critical_lower) & (x <= z_critical_upper)),
                   color='orange', alpha=0.5, label=f'Type II Error (β = {beta:.3f})')

  # Plot critical values
  plt.axvline(z_critical_lower, color='black', linestyle='--', label=f'Critical Value Lower')
  plt.axvline(z_critical_upper, color='black', linestyle='--', label=f'Critical Value Upper')

  plt.title("Visualization of Type I and Type II Errors")
  plt.xlabel("Sample Mean")
  plt.ylabel("Probability Density")
  plt.legend()
  plt.grid(True)
  plt.show()

  print(f"\n--- Error Probabilities ---")
  print(f"Alpha (Type I Error Probability): {alpha:.3f}")
  print(f"Beta (Type II Error Probability): {beta:.3f}")
  print(f"Power (1 - Beta)              : {1 - beta:.3f}")


# --- Example Usage ---
# Parameters for visualization
population_mean_h0 = 100  # Mean under the null hypothesis
true_mean_h1 = 105       # True mean (assuming H1 is true)
population_std_dev = 15    # Population standard deviation
sample_size = 30         # Sample size
significance_level = 0.05 # Alpha

visualize_errors(population_mean_h0, true_mean_h1, population_std_dev, sample_size, significance_level)

# Another example with different parameters
visualize_errors(mu0=50, mu1=52, sigma=5, n=50, alpha=0.01)
```

# 6. Write a Python program to perform an independent T-test and interpret the results.

```python
import numpy as np
from scipy import stats

# --- Input Data ---
# Sample data for two independent groups
# Example: Scores on a test for two different teaching methods
group1_scores = [85, 88, 90, 82, 87, 91, 84, 86, 89, 83]
group2_scores = [78, 80, 83, 75, 79, 82, 77, 81, 80, 76]

# Significance level (alpha)
alpha = 0.05

# --- Perform Independent T-test ---
# scipy.stats.ttest_ind performs an independent two-sample t-test.
# equal_var=True assumes equal population variances (Student's t-test).
# Set equal_var=False if variances are unequal (Welch's t-test).
t_statistic, p_value = stats.ttest_ind(group1_scores, group2_scores, equal_var=True)

# --- Display Results ---
print(f"Sample Size Group 1: {len(group1_scores)}")
print(f"Sample Size Group 2: {len(group2_scores)}")
print(f"Mean Group 1       : {np.mean(group1_scores):.2f}")
print(f"Mean Group 2       : {np.mean(group2_scores):.2f}")
print("-" * 30)
print(f"T-Statistic        : {t_statistic:.3f}")
print(f"P-Value            : {p_value:.4f}")
print(f"Significance Level : {alpha}")
print("-" * 30)

# --- Interpretation ---
print("Interpretation:")
if p_value < alpha:
    print(f"Since the P-value ({p_value:.4f}) is less than the significance level ({alpha}),")
    print("we reject the null hypothesis.")
    print("Conclusion: There is a statistically significant difference between the means of the two groups.")
else:
    print(f"Since the P-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}),")
    print("we fail to reject the null hypothesis.")
    print("Conclusion: There is no statistically significant difference between the means of the two groups.")

# Optional: Print means to see which group has a higher mean if difference is significant
if p_value < alpha:
    if np.mean(group1_scores) > np.mean(group2_scores):
        print("\nGroup 1 has a significantly higher mean than Group 2.")
    else:
        print("\nGroup 2 has a significantly higher mean than Group 1.")
```

# 8. Simulate data and perform both Z-test and T-test, then compare the results using Python.

```python
import numpy as np
from scipy.stats import norm, t

np.random.seed(0)  # Reproducibility
mu_true = 100       # True population mean
sigma_true = 15     # True population std dev
n = 20              # Sample size

sample = np.random.normal(loc=mu_true, scale=sigma_true, size=n)
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # Sample standard deviation (ddof=1 for unbiased)
mu_hypothesis = 100                  # Null hypothesis mean
alpha = 0.05                         # Significance level

z_score = (sample_mean - mu_hypothesis) / (sigma_true / np.sqrt(n))
p_value_z = 2 * (1 - norm.cdf(abs(z_score)))

t_score = (sample_mean - mu_hypothesis) / (sample_std / np.sqrt(n))
df = n - 1
p_value_t = 2 * (1 - t.cdf(abs(t_score), df))

print("===== Z-test (σ known) =====")
print(f"Z-Score      : {z_score:.3f}")
print(f"P-Value (Z)  : {p_value_z:.4f}")
print(f"Decision     : {'Reject' if p_value_z < alpha else 'Fail to Reject'} H₀")

print("\n===== T-test (σ unknown) =====")
print(f"T-Score      : {t_score:.3f}")
print(f"P-Value (T)  : {p_value_t:.4f}")
print(f"Degrees of freedom: {df}")
print(f"Decision     : {'Reject' if p_value_t < alpha else 'Fail to Reject'} H₀")

print("\nSummary:")
print("Z-test assumes known population standard deviation.")
print("T-test uses sample standard deviation, so it’s more conservative for small samples.")
```

# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy.stats import norm, t

def confidence_interval_mean(sample, confidence=0.95, sigma=None):

    sample = np.array(sample)
    n = len(sample)
    mean = np.mean(sample)

    if sigma is not None:
        # Use Z-distribution
        se = sigma / np.sqrt(n)
        z = norm.ppf((1 + confidence) / 2)
        margin = z * se
        dist = "Z"
    else:
        # Use T-distribution
        s = np.std(sample, ddof=1)
        se = s / np.sqrt(n)
        t_val = t.ppf((1 + confidence) / 2, df=n - 1)
        margin = t_val * se
        dist = "T"

    lower, upper = mean - margin, mean + margin
    return {
        "mean": mean,
        "lower": lower,
        "upper": upper,
        "confidence": confidence,
        "distribution": dist,
        "margin_of_error": margin,
        "std_error": se,
        "n": n
    }

# Example Usage:
sample_data = [12.7, 11.9, 13.2, 12.5, 12.0, 13.0, 12.3]
result = confidence_interval_mean(sample_data, confidence=0.95)

print(f"Sample Mean        : {result['mean']:.2f}")
print(f"{int(result['confidence']*100)}% CI using {result['distribution']}-distribution:")
print(f"  Lower Bound       : {result['lower']:.2f}")
print(f"  Upper Bound       : {result['upper']:.2f}")
print(f"Margin of Error    : {result['margin_of_error']:.2f}")
print(f"Standard Error (SE): {result['std_error']:.2f}")
print(f"Sample Size        : {result['n']}")
```

# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

```python
import numpy as np
from scipy.stats import t

def margin_of_error(sample, confidence=0.95):

    sample = np.array(sample)
    n = len(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (unbiased)
    se = sample_std / np.sqrt(n)          # Standard error

    # Degrees of freedom
    df = n - 1

    # Critical t value
    t_crit = t.ppf((1 + confidence) / 2, df)

    # Margin of error
    moe = t_crit * se
    return moe

# Example usage:
sample_data = [12.7, 11.9, 13.2, 12.5, 12.0, 13.0, 12.3]
confidence_level = 0.95

moe = margin_of_error(sample_data, confidence=confidence_level)
print(f"Margin of Error at {int(confidence_level*100)}% confidence: {moe:.3f}")
```

# 11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayes_theorem(prior_H, likelihood_E_given_H, prior_not_H, likelihood_E_given_not_H):

    # Calculate total probability of evidence P(E)
    prob_E = (likelihood_E_given_H * prior_H) + (likelihood_E_given_not_H * prior_not_H)

    # Apply Bayes' Theorem
    posterior_H = (likelihood_E_given_H * prior_H) / prob_E
    return posterior_H

prior_H = 0.01
prior_not_H = 1 - prior_H
likelihood_E_given_H = 0.9
likelihood_E_given_not_H = 0.05

posterior = bayes_theorem(prior_H, likelihood_E_given_H, prior_not_H, likelihood_E_given_not_H)

print(f"Posterior probability patient has disease given positive test: {posterior:.4f}")
```

```python
import numpy as np
from scipy.stats import chi2_contingency

data = np.array([[30, 10],
                 [20, 40]])

# Perform Chi-square test
chi2_stat, p_value, dof, expected = chi2_contingency(data)

print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"Degrees of Freedom : {dof}")
print(f"P-value             : {p_value:.4f}")
print("Expected Frequencies:")
print(expected)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis - variables are dependent (associated).")
else:
    print("Fail to reject null hypothesis - variables are independent.")
```

# 13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

```python
import numpy as np

def calculate_expected_frequencies(observed):

    observed = np.array(observed)
    row_totals = observed.sum(axis=1).reshape(-1, 1)
    col_totals = observed.sum(axis=0).reshape(1, -1)
    total = observed.sum()

    expected = (row_totals @ col_totals) / total
    return expected

# Example observed data: 2x3 contingency table
observed_data = [
    [30, 10, 20],
    [20, 40, 30]
]

expected_freq = calculate_expected_frequencies(observed_data)

print("Observed Frequencies:")
print(np.array(observed_data))
print("\nExpected Frequencies:")
print(np.round(expected_freq, 2))
```

# 14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

```python
import numpy as np
from scipy import stats

observed_counts = np.array([45, 55, 52, 48]) # e.g., counts in 4 categories

expected_proportions = np.array([0.25, 0.25, 0.25, 0.25])

total_observed = observed_counts.sum()

# Calculate expected counts based on expected proportions
expected_counts = expected_proportions * total_observed

# Ensure expected counts are not zero (requirement for Chi-square test)
if np.any(expected_counts <= 0):
    print("Error: Expected counts must be positive for the Chi-square test.")
else:
    # Calculate Chi-square statistic
    chi2_stat, p_value = stats.chisquare(f_obs=observed_counts, f_exp=expected_counts)

    # Degrees of freedom for goodness-of-fit is k - 1, where k is the number of categories
    dof = len(observed_counts) - 1

    # --- Display Results ---
    print("--- Goodness-of-Fit Test Results ---")
    print(f"Observed Counts     : {observed_counts}")
    print(f"Expected Counts     : {expected_counts.round(2)}")
    print("-" * 35)
    print(f"Chi-square Statistic: {chi2_stat:.4f}")
    print(f"Degrees of Freedom  : {dof}")
    print(f"P-value             : {p_value:.4f}")
    print("-" * 35)

    # --- Interpretation ---
    alpha = 0.05 # Significance level
    print("Interpretation:")
    if p_value < alpha:
        print(f"Since the P-value ({p_value:.4f}) is less than the significance level ({alpha}),")
        print("we reject the null hypothesis.")
        print("Conclusion: The observed data distribution is significantly different from the expected distribution.")
    else:
        print(f"Since the P-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}),")
        print("we fail to reject the null hypothesis.")
        print("Conclusion: The observed data distribution is not significantly different from the expected distribution.")
```
